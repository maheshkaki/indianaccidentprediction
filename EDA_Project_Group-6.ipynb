##### Replacing Missing Value ( Data fill and Interpolate)

```python
# # There are drivers who dont have Driving License and their field is mentioned as None.
# # So replace the None value to Driving Liscense not Issued
print(df[['Driver License Status','Year']].head())
df['Driver License Status'] = df['Driver License Status'].fillna('Driving Liscense not Issued')
df[['Driver License Status','Year']].head()
```

##### Replacing false data 'Pedestrain' in column 'Vehicle Type Involved' with ffill/bfill after making it None

```python
df['Vehicle Type Involved'].unique() # printing the uniques list of Vechicles involved in the accidents
```

```python
# From the data we have seen that 'Pedestrians' cant be a vehicle involved so replacing it with None
df['Vehicle Type Involved']=df['Vehicle Type Involved'].replace('Pedestrian',None) # Replacing the Pedestrian with None
df['Vehicle Type Involved'].unique() # Now the unique will have None vehicle Type
```

##### Now Interpolate the data based on 'Number of Casualties' for filling None values in 'Vehicle Type Involved'

```python
df['Vehicle Type Involved'] = df.groupby('Number of Casualties')['Vehicle Type Involved'].apply(lambda group: group.ffill().bfill()).reset_index(drop=True) # # Now None value is replaced by using 'Number of Casualties' using groupby
df['Vehicle Type Involved'].unique() ## view the unique values none is replaced.
```

##### Updating Unknown values in column 'City Name' to Capital city of States using Value Mapping

```python

# As the city Name column is unknown for many cells, we are filling the unknown with State Capital city name. Below is the Custom data for filling the dataframe.
state_cities = { "Jammu and Kashmir": "Srinagar", "Sikkim":"Gongtok", "Meghalaya":"Shillong", "Assam":"Dispur","Telangana":"Hyderabad","Madhya Pradesh":"Indore",
                "Odisha":"Bhuvaneshwar", "Delhi":"New Delhi"} 

df.loc[df['City Name'] == 'Unknown', 'City Name'] = df.loc[df['City Name'] == 'Unknown', 'State Name'].map(state_cities) # 
df['City Name'].unique() # Unknown values are filled with custom values
```

#### Number of casulaties should be Greater than equal to Number of fatalities

```python
df = df.reset_index(drop=True) # Reset the index to ensure sequential numeric indices
# This step is essential when working with row-wise updates to avoid KeyError issues.
values = [(i, val) for i, val in enumerate(df['Number of Fatalities']) if df['Number of Casualties'][i] < val] # Create a list of tuples containing row index and the value to be updated in 'Number of Casualties'
# Only rows where 'Number of Casualties' is less than 'Number of Fatalities' are considered.

for i, val in values: # Iterate over the list of values and update the 'Number of Casualties' column for each row
    df.loc[i, 'Number of Casualties'] = val
    
df[['Number of Casualties', 'Number of Fatalities']]
```

## Groupby function 

#### Group by severity of accident 

```python
# Groups the DataFrame `df` by the 'Accident Severity' column.
# Computes the mean of the columns ['Number of Vehicles Involved', 'Number of Casualties', 'Number of Fatalities'] for each severity group.
# Rounds the mean values to 2 decimal places for better readability.
# Resets the index to turn the 'Accident Severity' grouping into a regular column in the resulting DataFrame.
average_accident_severity=df.groupby('Accident Severity')[['Number of Vehicles Involved','Number of Casualties','Number of Fatalities']].mean().round(2).reset_index()
# Renames the columns of the resulting DataFrame for clarity.
# The column names now clearly represent the average values, with the 'AVG_' prefix indicating "average."
average_accident_severity.columns=['Accident Severity','AVG_Number of Vehicles Involved','AVG_Number of Casualties','AVG_Number of Fatalities']
# Applies styling to the DataFrame for better presentation.
# Formats the numeric values in the specified columns to display 2 decimal places, making the output more visually appealing.
# Uses `subset` to target only the columns with average values for formatting.
average_accident_severity.style \
  .format("{:.2f}",subset=['AVG_Number of Vehicles Involved', 'AVG_Number of Casualties', 'AVG_Number of Fatalities']) 
```

### Group by weather condition

```python
# Groups the DataFrame `df` by the 'Weather Conditions' column.
# Calculates the mean for the specified columns: 'Number of Vehicles Involved', 'Number of Casualties', 'Number of Fatalities'.
# The result is a DataFrame where each row corresponds to a unique weather condition and the averaged values of the numeric columns.
# The `.reset_index()` converts the grouped DataFrame back into a regular DataFrame structure for further processing or display.
average_weather_condition=df.groupby('Weather Conditions')[['Number of Vehicles Involved','Number of Casualties','Number of Fatalities']].mean().reset_index()
# Renames the columns of the resulting DataFrame for clarity.
# The column names now explicitly show that they contain average values (using the prefix 'AVG_').
average_weather_condition.columns=['Weather Conditions','AVG_Number of Vehicles Involved','AVG_Number of Casualties','AVG_Number of Fatalities']
# Applies styling to the DataFrame for better presentation, typically used in environments like Jupyter Notebooks.
# Formats the numeric values in the specified columns to display 2 decimal places, improving readability.
# The `subset` parameter ensures the formatting is applied only to the relevant numeric columns (average values).
average_weather_condition.style.format("{:.2f}",subset=['AVG_Number of Vehicles Involved','AVG_Number of Casualties','AVG_Number of Fatalities'])
```

### Group by road type


```python
# Groups the DataFrame `df` by the 'Road Type' column, creating groups based on unique values in 'Road Type'.
# Computes the mean of the specified numeric columns ('Number of Vehicles Involved', 'Number of Casualties', 'Number of Fatalities') for each road type.
# The `.reset_index()` converts the grouped DataFrame back into a regular DataFrame structure by restoring 'Road Type' as a column instead of an index.
average_road_type=df.groupby('Road Type')[['Number of Vehicles Involved','Number of Casualties','Number of Fatalities']].mean().reset_index()
# Renames the columns of the resulting DataFrame for clarity.
# Adds the 'AVG_' prefix to the numeric columns to indicate they contain average values, making the DataFrame easier to interpret.
average_road_type.columns=['Road Type','AVG_Number of Vehicles Involved','AVG_Number of Casualties','AVG_Number of Fatalities']
# Applies formatting to the DataFrame for presentation purposes, especially in environments like Jupyter Notebook.
# Formats numeric values in the specified columns to display 2 decimal places, ensuring consistency and readability.
# The `subset` parameter targets only the columns with average values for formatting, leaving other columns unchanged.
average_road_type.style.format('{:.2f}',subset=['AVG_Number of Vehicles Involved','AVG_Number of Casualties','AVG_Number of Fatalities'])
```

##### Using user defined function and Lambda fnction to create a column "Date_of_incident" by calcuating 'year', 'month' and 'Day of week'

```python
# Function to get the first occurrence of a weekday in a given month and year
def get_first_weekday(year, month, weekday_name):
    # Mapping month names with their numerical representation
    month_map = {name: i for i, name in enumerate(calendar.month_name) if name} 
    month_num = month_map[month] # Month in numeric value is assigned to this variable
    
    # Mapping weekday names to their integer representation
    weekday_map = {name: i for i, name in enumerate(calendar.day_name)}
    weekday = weekday_map[weekday_name] # Day in numeric value is assigned to this variable
    
    # Get the first day of the month
    first_day_of_month = pd.Timestamp(year, month_num, 1)
    
    # Calculating the first appearing date of the specified weekday
    days_to_add = (weekday - first_day_of_month.weekday() + 7) % 7
    first_weekday_date = first_day_of_month + pd.Timedelta(days=days_to_add)
    return first_weekday_date

# Using function a new column of a dataframe is be updated for easier analysis 
df["Date_of_incident"] = df.apply(lambda row: get_first_weekday(row['Year'], row['Month'], row['Day of Week']), axis=1)
df[['State Name', 'Year', 'Accident Location Details', 'Date_of_incident']].head()

```
